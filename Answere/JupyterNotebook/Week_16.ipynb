```
import qiskit as q
import numpy as np
import matplotlib.pyplot as plt
from qiskit import Aer, execute, assemble
from qiskit.visualization import plot_histogram
import math
%matplotlib inline
def measure(aer_backend,circuit):
    backend = q.Aer.get_backend(aer_backend)
    job = q.execute(circuit, backend)
    result = job.result()
    counts = result.get_counts(circuit)
    graph = q.visualization.plot_histogram(counts)
    display(graph)
    
def draw():
    display(circuit.draw(output="mpl"))

def draw_prob(circuit):
    backend = q.Aer.get_backend('statevector_simulator')
    job = execute(circuit,backend)
    result = job.result()
    counts = result.get_counts(circuit)
    graph=plot_histogram(counts)
    display(graph)
circuit = q.QuantumCircuit(2)
draw()
```

```
circuit.h(0)
circuit.cnot(0,1)
draw()
```

```
measure('statevector_simulator',circuit)
```
